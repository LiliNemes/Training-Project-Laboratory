```python
df_for_netw = pd.DataFrame(columns=['Source', 'Target', 'Weight'])

df_for_netw.head(5)
```

```python
for i in range(0, len(expr_df_shorter)):
  for j in range(0, len(expr_df_shorter)):
    if (coexpression_matrix[i][j]>0) and (i!=j):
      df_for_netw.loc[len(df_for_netw.index)] = [labels_dict[short_index[i]], labels_dict[short_index[j]], 1]
  is_there = False
  for z in range(0, len(df_for_netw.index)):
    if(df_for_netw.Source[z]==labels_dict[short_index[i]]):
      is_there=True
  if(is_there==False):
    df_for_netw.loc[len(df_for_netw.index)]=[labels_dict[short_index[i]],labels_dict[short_index[i]], 0]

df_for_netw.head(30)
```

```python
import networkx as nx
```

```python
G = nx.from_pandas_edgelist(df_for_netw, source='Source', target='Target', edge_attr='Weight')
```
